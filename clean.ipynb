```python
import pandas as pd
```

```python
df = pd.read_excel('frederick_2005.xls')
df = df.dropna(subset=['ACT', 'RACE', '3.4K v 3.8K(2)', 'Sex', 'IQ'])
df = df.dropna(axis=1)
df = df[[
    'school',
    'Sex',
    'RACE',
    'ACT',
    'batball',
    '3flips',
    'widgets',
    'trucks',
    'lilypads',
    'water',
    'giftwrap',
    'Jerry',
    'IQ',
    '3.4K v 3.8K(2)',
    'Prestige',
]]
df.head()
```

```python
df.to_csv('CRT_nodummies.csv', index=False)
```

```python
df = pd.get_dummies(df, columns=['school', 'Sex', 'RACE', 'Prestige'])
df['CRT'] = df[['batball', 'widgets', 'lilypads']].sum(axis=1)
df = df.drop(columns='Sex_Female')
df.head()
```

```python
df.to_csv('CRT_clean.csv', index=False)
```
